```python
import pandas as pd
```

```python
housing= pd.read_csv("price.csv")
```

```python
housing.info()
```

```python
housing.describe()
```

```python
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
#line plot
housing.area.plot(kind='line', color = 'g', label= 'area', linewidth= 1, alpha =0.5, grid = True, linestyle = '-')
housing.price.plot(color = 'r',label = 'price', linewidth=1, alpha= 0.5,grid= True,linestyle='-.')
plt.legend(loc='upper right')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Line Plot')
plt.show()
            
```

```python
#scatter plot
#x = area, y= price
housing.plot(kind='scatter',x='area', y = 'price',alpha = 0.5,color = 'red')
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Scatter Plot')
```

```python
#histogram
housing.price.plot(kind = 'hist',bins= 60, figsize = (20,20))
plt.show()
```
